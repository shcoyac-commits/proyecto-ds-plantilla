```python
import numpy as np

# --- S2.3 Numpy esencial: arrays, broadcasting, vectorización vs loops ---
# Precios de ensaladas vendidas en un día
precios_lista = [45.50, 62.00, 38.75, 55.20, 71.00]

# Convertir a array de NumPy
precios = np.array([45.50, 62.00, 38.75, 55.20, 71.00])

print(type(precios_lista))
print(type(precios))
print(precios)
```

```python
# Operaciones vectorizadas — sin loops
precios_con_iva = precios * 1.16
descuento_10 = precios * 0.90
ganancia = precios - 28  # costo base de $28 por ensalada

print("Con IVA:     ", precios_con_iva)
print("Con 10% desc:", descuento_10)
print("Ganancia:    ", ganancia)
```

```python
# Broadcasting — operar arrays de diferente forma
# Imagina 3 sucursales, cada una vendió 5 ensaladas
ventas = np.array([
    [45.50, 62.00, 38.75, 55.20, 71.00],  # Sucursal A
    [39.00, 58.50, 42.00, 61.00, 53.75],  # Sucursal B
    [52.00, 47.25, 66.00, 44.50, 59.80],  # Sucursal C
])

print("Forma del array:", ventas.shape)  # (3, 5) = 3 filas, 5 columnas

# Broadcasting: multiplicar toda la tabla por un escalar
print("\nCon IVA (todo):\n", ventas * 1.16)

# Broadcasting: cada sucursal tiene un factor de ajuste distinto
ajuste_sucursal = np.array([[1.10], [1.00], [0.95]])  # A sube 10%, B igual, C baja 5%
print("\nForma del ajuste:", ajuste_sucursal.shape)  # (3, 1)

ventas_ajustadas = ventas * ajuste_sucursal
print("\nVentas ajustadas:\n", ventas_ajustadas)
```

```python
# 10 ventas x 5 productos
ventas_grande = np.random.randint(20, 80, size=(10, 5))
print("Ventas:", ventas_grande.shape)      # (10, 5)

# Caso 1: ajuste (3, 1) — NO es compatible con (10, 5)
ajuste_mal = np.array([[1.10], [1.00], [0.95]])
print("Ajuste malo:", ajuste_mal.shape)    # (3, 1)

try:
    resultado = ventas_grande * ajuste_mal
except ValueError as e:
    print("ERROR:", e)

# Caso 2: ajuste (10, 1) — SÍ es compatible
ajuste_bien = np.random.uniform(0.90, 1.10, size=(10, 1))
print("\nAjuste bueno:", ajuste_bien.shape)  # (10, 1)
resultado = ventas_grande * ajuste_bien
print("Resultado:", resultado.shape)        # (10, 5) ✓

# Caso 3: ajuste (1, 5) — también compatible, ajusta por columna
ajuste_por_producto = np.array([[1.16, 1.08, 1.00, 0.95, 1.20]])
print("\nAjuste por producto:", ajuste_por_producto.shape)  # (1, 5)
resultado2 = ventas_grande * ajuste_por_producto
print("Resultado:", resultado2.shape)       # (10, 5) ✓
```

```python
# Retomamos las 3 sucursales con 5 ensaladas cada una
ventas = np.array([
    [45.50, 62.00, 38.75, 55.20, 71.00],  # Sucursal A
    [39.00, 58.50, 42.00, 61.00, 53.75],  # Sucursal B
    [52.00, 47.25, 66.00, 44.50, 59.80],  # Sucursal C
])

# axis=0 → colapsa las filas, te da UN valor por columna
print("Promedio por producto (axis=0):", ventas.mean(axis=0))  # 5 valores

# axis=1 → colapsa las columnas, te da UN valor por fila
print("Promedio por sucursal (axis=1):", ventas.mean(axis=1))  # 3 valores

# Sin axis → promedio de TODOS los elementos
print("Promedio general:", ventas.mean())
```

```python
# np.where — como un IF de Excel
# Clasificar: si el precio es mayor a 55, es "Premium", si no es "Regular"
precios = np.array([45.50, 62.00, 38.75, 55.20, 71.00])

categoria = np.where(precios > 55, "Premium", "Regular")
print("np.where:", categoria)

# np.select — como un IF anidado / SWITCH
# Múltiples condiciones: Económica (<40), Regular (40-60), Premium (>60)
condiciones = [
    precios < 40,
    (precios >= 40) & (precios <= 60),
    precios > 60
]
etiquetas = ["Económica", "Regular", "Premium"]

segmento = np.select(condiciones, etiquetas, default="Sin clasificar")
print("np.select:", segmento)
```

```python
import timeit

# Simular 1 millón de tickets de venta
np.random.seed(42)
precios_grandes = np.random.uniform(25, 85, size=1_000_000)

# Método 1: loop puro de Python
def calcular_iva_loop(precios):
    resultado = []
    for p in precios:
        resultado.append(p * 1.16)
    return resultado

# Método 2: operación vectorizada con NumPy
def calcular_iva_numpy(precios):
    return precios * 1.16

# Medir tiempos
tiempo_loop = timeit.timeit(lambda: calcular_iva_loop(precios_grandes), number=5)
tiempo_numpy = timeit.timeit(lambda: calcular_iva_numpy(precios_grandes), number=5)

print(f"Loop Python:  {tiempo_loop:.4f} segundos (5 ejecuciones)")
print(f"NumPy vector: {tiempo_numpy:.4f} segundos (5 ejecuciones)")
print(f"\nNumPy es {tiempo_loop / tiempo_numpy:.0f}x más rápido")
```

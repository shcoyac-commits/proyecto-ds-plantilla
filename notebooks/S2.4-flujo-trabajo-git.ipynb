## Flujo de trabajo Git para cada sesión

### Antes de empezar a trabajar
1. Abrir terminal en VSCode
2. Activar ambiente: `conda activate analytics`
3. Verificar rama actual: `git branch`
4. Actualizar desde remoto: `git pull`

### Durante la sesión
- Guardar notebooks frecuentemente (Ctrl+S)
- Los notebooks se auto-guardan, pero es buena práctica hacerlo manual

### Al final de cada sesión
1. Revisar cambios: `git status`
2. Agregar archivos: `git add .`
3. Commit con mensaje descriptivo: `git commit -m "Descripción clara del trabajo"`
4. Push a GitHub: `git push`

### Buenas prácticas
- **Un commit por sesión** con mensaje que indique el módulo (ej: "S2.4 Flujo de trabajo Git")
- **Nunca commitear** archivos en `.ipynb_checkpoints/` (ya están en .gitignore)
- **Validar** que el push fue exitoso viendo el mensaje "main -> main"

```python
import pandas as pd
import hashlib

# Cargar dataset
df = pd.read_parquet(r'C:\Users\Said\Documents\Datasets\Alimentos\Dataset1.parquet')

print(f"Shape original: {df.shape}")

# 1. ELIMINAR NombreCliente2
df = df.drop(columns=['NombreCliente2'])
print("✓ Eliminada columna NombreCliente2")

# 2. HASHEAR IdCliente con salt
SALT = "leonali_2024_secret"  # ⚠️ GUARDAR EN .env, NO EN GITHUB
def hash_id(value, salt=SALT):
    return hashlib.sha256(f"{value}{salt}".encode()).hexdigest()[:12]

df['IdCliente'] = df['IdCliente'].apply(hash_id)
print("✓ Hasheado IdCliente")

# 3. CONVERTIR FechaProceso a datetime (mantener completa, SIN redondear)
df['FechaProceso'] = pd.to_datetime(df['FechaProceso'], format='%d/%m/%Y')
print("✓ Convertida FechaProceso a datetime (granularidad completa)")

print(f"\nShape final: {df.shape}")
print("\nPrimeras 3 filas anonimizadas:")
print(df.head(3))

# 4. GUARDAR dataset anonimizado
df.to_parquet(r'C:\Users\Said\Documents\Datasets\Alimentos\Dataset1_anonimizado.parquet', index=False)
print(f"\n✓ Dataset anonimizado guardado")
print(f"✓ Tamaño: {df.memory_usage(deep=True).sum() / 1e6:.2f} MB")

print("\n" + "="*60)
print("RESUMEN DE ANONIMIZACIÓN:")
print("="*60)
print("✓ NombreCliente2: ELIMINADA")
print("✓ IdCliente: HASHEADO (con salt secreto)")
print("✓ FechaProceso: DATETIME completa (sin redondeo)")
print("✓ Resto de columnas: MANTENIDAS")
```
